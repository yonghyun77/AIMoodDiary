## load data

```python
with open('sentence_for_preprocessing_package_choice.txt', mode='rt', encoding='utf-8') as f: # rt(텍스트 읽기모드), f라는 변수에 파일 열기
  lines = f.readlines() # 모든 줄을 lines에 저장
```

```python
lines
```

```python
data = []

for line in lines: # 한 줄씩
  line = line.rstrip('\n') # 개행문자('\n') 제거
  line = line.strip() # 앞, 뒤 공백 제거
  data.append(line[5:]) # 감정은 제외하고 문장만 data에 저장

data
```

## morpheme analysis-  mecab

```python
from konlpy.tag import Mecab

mcb = Mecab()
```

```python
mecab_lst = []

for d in data:
  for word in mcb.pos(d):
    mecab_lst.append(word)

mecab_lst
```

## morpheme analysis - khaiii

```python
from khaiii import KhaiiiApi

api = KhaiiiApi()
```

```python
def to_tuple(a, b): # khaiii는 결과가 튜플이 아니라서 튜플로 저장하기 위한 함수 선언
  return (a, b)
```

```python
khaiii_lst = []

for d in data:
  for word in api.analyze(d):
    for morph in word.morphs:
      khaiii_lst.append(to_tuple(morph.lex, morph.tag))

khaiii_lst
```

## tuple to dictionary

형태소가 영어로 되어 있으면 알아보기 힘들 것 같아서 한글로 바꾸고 싶은데 튜플은 값을 수정할 수 없기 때문에 딕셔너리 형태로 변경


### mecab

```python
mecab_dict = {}
for word in mecab_lst:
  mecab_dict.update({word[0]: word[1]})
```

```python
mecab_dict
```

### khaiii

```python
khaiii_dict = {}
for word in khaiii_lst:
  khaiii_dict.update({word[0]: word[1]})
```

```python
khaiii_dict
```

## eng to kor

### macab

```python
mecab_dict_ko = {}

for w, m in mecab_dict.items():
  for key, value in mcb.tagset.items(): # mecab에서 제공하는 형태소 사전 활용
    if m == key:
      mecab_dict_ko.update({w : value})
    
```

```python
mecab_dict_ko
```

```python
for d in data:
  for word in mcb.pos(d):
    if word[0] in mecab_dict_ko.keys():
      continue
    else:
      mecab_dict_ko.update({word[0] : word[1]})
      # mecab의 경우 NNG+JX 등 두 개의 형태소가 합쳐진 형태가 등장, 이 경우 영어 형태소 그대로 넣어줌
```

```python
mecab_dict_ko
```

### khaiii

```python
khaiii_dict_ko = {}

for w, m in khaiii_dict.items():
  for key, value in mcb.tagset.items(): 
    # khaiii는 별도의 형태소 사전을 제공하지 않는 것 같아서(이미지로는 있는데 ..) mecab의 형태소 사전 활용 (형태소 사전은 다 똑같겠지 뭐)
    if m == key:
      khaiii_dict_ko.update({w : value})
```

```python
khaiii_dict_ko
```

## compare

```python
mecab_result = []
s = ''
for d in data:
  for word in mcb.pos(d):
    for key, value in mecab_dict_ko.items():
      if word[0] == key:
        print(f'{word[0]}({value})', end=' ')
        s += word[0] + '(' + value + ') '
  mecab_result.append(s) # mecab랑 khaiii랑 한 문장씩 비교하기 위해 list에 저장
  s = ''
```

```python
khaiii_result = []
s = ''
for d in data:
  for word in api.analyze(d):
    for morph in word.morphs:
      for key, value in khaiii_dict_ko.items():
        if morph.lex == key:
          print(f'{morph.lex}({value})', end=' ')
          s += morph.lex + '(' + value + ') '
  khaiii_result.append(s)
  s = ''
```

```python
for i in range(len(data)):
  print(f'M: {mecab_result[i]}')
  print(f'K: {khaiii_result[i]}')
  print()
```
